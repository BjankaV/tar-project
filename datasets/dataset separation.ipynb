```python
import pandas as pd
```

```python
tsv_file='OLIDv1.0/olid-training-v1.0.tsv'
 
# reading given tsv file
csv_table=pd.read_table(tsv_file, sep='\t')
 
# converting tsv file into csv
csv_table.to_csv('train_all_levels.csv', index=False)
```

```python
df = pd.read_csv('train_all_levels.csv')
df
```

# Sub-task A

```python
df_a = df.copy()
df_a = df_a.drop(['subtask_b', 'subtask_c'], axis=1)
df_a
```

```python
df_a.to_csv('train_level_a.csv', index=False)
```

```python
df_a_small = df_a.head(3000)
df_a_small.to_csv('train_level_a_small.csv', index=False)
```
